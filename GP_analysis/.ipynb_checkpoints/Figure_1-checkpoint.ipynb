# Set up

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import GPy

plt.style.use('hfs')
```

```python
## loading data
results=pd.read_csv('params_best_GP_IIb.dat', index_col=0)
data_all = pd.read_csv('../Data/IIb_clean_sample.dat', index_col=0)
```

# SN 2008ax

```python
sn='SN2008ax'
data=data_all[data_all.event==sn].drop('event',axis=1)
```

```python
data.head()
```

## Train the GPs

### First the data is put into the right structure to be fed to the GP

```python
# We use the phase w.r.t. to epxlosion date so we correct our MJDs...
phase = data.MJD.values - data.texp.iloc[0]
t_exp = 0 # ...and set explosion time is set to 0

y=np.atleast_2d(data.V).T # dependent variable, here the magnitude
X=np.atleast_2d(phase).T # independent variable, here the phase 
```

```python
phase
```

### Then we set up the kernels
We are going to test the effect of three kernels here: the squared-exponential (called RBF), the white noise and the bias kernels. For the difference between the whtie noise and bias kernel see [this tutorial form the GPy team](https://nbviewer.org/github/SheffieldML/notebook/blob/master/GPy/basic_kernels.ipynb).


```python
# Square Exponential Kernel
length_scale=np.mean(np.gradient(phase)) # average of the time difference between exposures
variance=np.sqrt(np.std(np.gradient(phase)))

k_rbf = GPy.kern.RBF(input_dim=1, variance=variance, lengthscale=length_scale)
```

```python
length_scale, variance
```

```python
# White Noise Kernel
sigma = data.Verr.values

k_white = GPy.kern.White(1, GPy.util.linalg.tdot(sigma))
```

```python
# Bias Kernel
k_bias = GPy.kern.Bias(1, np.dot(sigma, sigma.T))
```

### Training/fitting

```python
# nnumber of restarts by the optimizer
n=100
```

```python
m_rbf = GPy.models.GPRegression(X,
                                y,
                                k_rbf)
__ = m_rbf.optimize ()
__ = m_rbf.optimize_restarts (num_restarts = n, 
                              parallel = True, 
                              verbose=False)
```

```python
m_rbf.plot(plot_density=True)
plt.gca().invert_yaxis()
plt.xlim([0,200])
plt.ylim([20,11])
plt.title('RBF only')
```

```python
m_rbf_white = GPy.models.GPRegression(X,
                                      y,
                                      k_rbf+k_white)
__ = m_rbf_white.optimize ()
__ = m_rbf_white.optimize_restarts (num_restarts = n, 
                                    parallel = True, 
                                    verbose=False)
```

```python
m_rbf_white.plot(plot_density=True)
plt.gca().invert_yaxis()
plt.xlim([0,200])
plt.ylim([20,11])
plt.title('RBF + White noise')
```

```python
m_rbf_bias = GPy.models.GPRegression(X,
                                     y,
                                     k_rbf+k_bias)
__ = m_rbf_bias.optimize ()
__ = m_rbf_bias.optimize_restarts (num_restarts = n, 
                                   parallel = True, 
                                   verbose=False)
```

```python
m_rbf_bias.plot(plot_density=True)
plt.gca().invert_yaxis()
plt.xlim([0,200])
plt.ylim([20,11])
plt.title('RBF + Bias')
```

### What do the noise kernels do?

To take a look at those noise kernels in the GP let's 'fit' the data just with just the white noise and the bias.

```python
m_white = GPy.models.GPRegression(X,
                                  y,
                                  k_white)
__ = m_white.optimize ()
__ = m_white.optimize_restarts (num_restarts = n, 
                                parallel = True, 
                                verbose=False)
```

```python
m_white.plot(plot_density=True)
plt.gca().invert_yaxis()
plt.xlim([0,200])
plt.ylim([20,-20])
plt.title('White only')
```

The white noise kernel is such that its mean is 0 and so its only way to try and 'fit' the data is to increase the variance around the mean. It will not affect the mean of the predictions.

```python
m_bias = GPy.models.GPRegression(X,
                                 y,
                                 k_bias)
__ = m_bias.optimize ()
__ = m_bias.optimize_restarts (num_restarts = n, 
                               parallel = True, 
                               verbose=False)
```

```python
m_bias.plot(plot_density=True)
plt.gca().invert_yaxis()
plt.xlim([0,200])
plt.ylim([20,12])
plt.title('Bias only')
```

The Bias kernel can move around the zero-point, it can take into account offsets as well as the stochastic white noise, meaning it affects the mean of the predicted data as well as their covariance. 


# Pretty plots

```python
fig = plt.figure(figsize=(12,7))

#gs = GridSpec(2,3, hspace=0.2)
#ax0 = fig.add_subplot(gs[0,:])
#ax1 = fig.add_subplot(gs[1, 0])
#ax2 = fig.add_subplot(gs[1, 1])
#ax3 = fig.add_subplot(gs[1, 2])


gs = GridSpec(3,2, hspace=0.2)
ax0 = fig.add_subplot(gs[:,0])
ax1 = fig.add_subplot(gs[0, 1])
ax2 = fig.add_subplot(gs[1, 1])
ax3 = fig.add_subplot(gs[2, 1])
ax=np.array([ax0, ax1, ax2, ax3])


## ==== PLOT 0 ===== 

## - Observed Data -

ax[0].scatter(phase, data.V,  
              marker='+', color='k',  alpha=0.8, 
              lw=1.5, s=50,zorder=0.1, label='SN 2008ax')


## - Model Data -

# The new time array we want predictions for 
new_t = np.arange(0, phase.max(),0.1)

# now because of the way the machine learning packages are written 
# they expect our data to be a vertical matrix, a.k.a  2D transposed version of our array 
# this gives our input dates the right shape
new_t = np.atleast_2d(new_t).T

V_rbf, __= m_rbf.predict(new_t)
V_rbf = V_rbf[:,0]

V_rbf_white, __= m_rbf_white.predict(new_t)
V_rbf_white = V_rbf_white[:,0]

V_rbf_bias, __= m_rbf_bias.predict(new_t)
V_rbf_bias = V_rbf_bias[:,0]

ax[0].plot(new_t, V_rbf, 
           label='$\mu$(t) RBF ', lw=2,
           alpha=0.9, color='cornflowerblue')

ax[0].plot(new_t, V_rbf_white, 
           label='$\mu$(t) RBF + White', lw=2,
           ls='--', alpha=0.7, color='blue')

ax[0].plot(new_t, V_rbf_bias, 
           label='$\mu$(t) RBF + Bias', lw=1,
           alpha=0.9, ls='-', color='crimson')

## === PLOT 1, 2, 3 ===

m_rbf.plot(plot_density=False, ax=ax[1], zorder=1000)
m_rbf_white.plot(plot_density=False, ax=ax[2], zorder=1000, legend=False)
m_rbf_bias.plot(plot_density=False, ax=ax[3], zorder=1000, legend=False)


## == AXES and LEGENDS ===

for axis in ax:
    axis.invert_yaxis()
    axis.set_ylim([24,10])
    
    axis.axvline(0, c='k', ls=':', alpha=0.3)
    
    
ax[0].legend()
ax[0].set_ylim([18,13])
ax[0].set_ylabel('V mag')
ax[0].text(1, 17.8, 'Explosion', color='k', alpha=0.5)
ax[0].set_xlabel('Phase (days)')

ax[1].text(40,12, 'kernel: RBF')
ax[1].text(1, 22, 'Explosion', color='k', alpha=0.5)

ax[2].text(40,12, 'kernel: RBF + white')
ax[2].text(1, 22, 'Explosion', color='k', alpha=0.5)

ax[3].text(40,12, 'kernel: RBF + bias')
ax[3].text(1, 22, 'Explosion', color='k', alpha=0.5)
ax[3].set_xlabel('Phase (days)')

#plt.savefig('sn2008ax.png', dpi=180, bbox_inches='tight')
```

```python
sn='SN2011dh'
data=data_all[data_all.event==sn].drop('event',axis=1)
data.head()
```

## Training the GPs

```python
# Square Exponential Kernel
length_scale=np.mean(np.gradient(phase)) # average of the time difference between exposures
variance=np.sqrt(np.std(np.gradient(phase)))
k_rbf = GPy.kern.RBF(input_dim=1, variance=variance, lengthscale=length_scale)

# White Noise Kernel
sigma = data.Verr.values
k_white = GPy.kern.White(1, GPy.util.linalg.tdot(sigma))

# Bias Kernel
k_bias = GPy.kern.Bias(1, np.dot(sigma, sigma.T))
```

```python
# We use the phase w.r.t. to epxlosion date so we correct our MJDs...
phase = data.MJD.values - data.texp.iloc[0]
t_exp = 0 # ...and set explosion time is set to 0

y=np.atleast_2d(data.V).T # dependent variable, here the magnitude
X=np.atleast_2d(phase).T # independent variable, here the phase 
```

```python
n=100
```

```python
m_rbf = GPy.models.GPRegression(X,
                                y,
                                k_rbf)
__ = m_rbf.optimize ()
__ = m_rbf.optimize_restarts (num_restarts = n, 
                              parallel = True, 
                              verbose=False)
```

```python
m_rbf.plot(plot_density=True)
plt.gca().invert_yaxis()
plt.xlim([0,90])
plt.ylim([17,11])
plt.title('RBF only')
```

```python
m_rbf_white = GPy.models.GPRegression(X,
                                      y,
                                      k_rbf+k_white)
__ = m_rbf_white.optimize ()
__ = m_rbf_white.optimize_restarts (num_restarts = n, 
                                    parallel = True, 
                                    verbose=False)
```

```python
m_rbf_white.plot(plot_density=True)
plt.gca().invert_yaxis()
plt.xlim([0,90])
plt.ylim([17,11])
plt.title('RBF + White noise')
```

```python
m_rbf_bias = GPy.models.GPRegression(X,
                                     y,
                                     k_rbf+k_bias)
__ = m_rbf_bias.optimize ()
__ = m_rbf_bias.optimize_restarts (num_restarts = n, 
                                   parallel = True, 
                                   verbose=False)
```

```python
m_rbf_bias.plot(plot_density=True)
plt.gca().invert_yaxis()
plt.xlim([0,90])
plt.ylim([17,11])
plt.title('RBF + Bias')
```

The likely reason the late time lightcurve is overfit is because the wiggly offset is being as a bias by the bias kernel not as stochastic noise, so it has affected the mean \mu of the result not just the covariance.
